```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.youtube.com/watch?v=9Mz84cwVmS0&ab_channel=Google'
```

```python
response = requests.get(url)
response
```

```python
raw_data = response.text
raw_data
```

```python
soup = BeautifulSoup(raw_data, 'html.parser')
soup
```

```python
tags = soup.find_all('a')
tags
```

```python
for tag in tags:
    print(tag.get('href'))
```
