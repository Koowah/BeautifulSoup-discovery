# BeautifulSoup introductory course : scrap all links from a page

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://boston.craigslist.org/search/sof"
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
data
```

```python
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
tags = soup.find_all('a')
tags
```

tags behaves like a list eventhough different type

to get the content of an element (<>content</>) use **.text**

to get the content of a field (< href='content'>...</>) of the element use **.get('subfield')**

```python
for tag in tags:
    print(tag.get('href'))
```

```python
titles = soup.find_all("a", {"class":"result-title"})
```

```python
for title in titles:
    print(title.text)
```

```python
locations = soup.find_all("span", {"class":"result-hood"})
```

```python
for location in locations:
    print(location.text)
```
